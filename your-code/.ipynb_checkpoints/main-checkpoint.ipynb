```python
import matplotlib.pyplot as plt

def simulate_dice_rolls(num_rolls=10):
    rolls = [random.randint(1, 6) for _ in range(num_rolls)]
    df = pd.DataFrame({'Roll': rolls})
    return df

# Simulate dice rolls and create DataFrame
dice_df = simulate_dice_rolls()

# Sort the DataFrame by the 'Roll' column
sorted_df = dice_df.sort_values(by='Roll')

# Plot the sorted results
plt.figure(figsize=(8, 6))
plt.bar(sorted_df.index, sorted_df['Roll'])
plt.xlabel('Roll Index')
plt.ylabel('Dice Value')
plt.title('Sorted Dice Rolls')
plt.xticks(sorted_df.index)
plt.grid(True)
plt.show()
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python
def simulate_dice_rolls(num_rolls=10):
    rolls = [random.randint(1, 6) for _ in range(num_rolls)]
    df = pd.DataFrame({'Roll': rolls})
    return df

dice_df = simulate_dice_rolls()

frequency_distribution = dice_df['Roll'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
plt.bar(frequency_distribution.index, frequency_distribution.values)
plt.xlabel('Dice Value')
plt.ylabel('Frequency')
plt.title('Dice Roll Frequency Distribution')
plt.xticks(range(1, 7))
plt.grid(True)
plt.show()
```

```python
"""
the first plot gives a visual representation of the sequence of rolls, 
while the second plot gives a summary of how often each value occurred in the rolls.
"""
```

## Challenge 2
Now, using the dice results obtained in *challenge 1*, your are going to define some functions that will help you calculate the mean of your data in two different ways, the median and the four quartiles. 

#### 1.- Define a function that computes the mean by summing all the observations and dividing by the total number of observations. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
def calculate_mean(df):
    total_sum = 0
    num_observations = len(df)

    for value in df['Roll']:
        total_sum += value

    mean = total_sum / num_observations
    return mean

mean = calculate_mean(dice_df)
print("Mean:", mean)
```

#### 2.- First, calculate the frequency distribution. Then, calculate the mean using the values of the frequency distribution you've just computed. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
def calculate_mean_from_frequency(frequency_distribution):
    total_sum = 0
    total_observations = 0

    for value, frequency in frequency_distribution.items():
        total_sum += value * frequency
        total_observations += frequency

    mean = total_sum / total_observations
    return mean

mean_from_frequency = calculate_mean_from_frequency(frequency_distribution)
print("Mean:", mean_from_frequency)
```

#### 3.- Define a function to calculate the median. You are not allowed to use any methods or functions that directly calculate the median value. 
**Hint**: you might need to define two computation cases depending on the number of observations used to calculate the median.

```python
def calculate_median(df):
    sorted_values = sorted(df['Roll'])
    num_observations = len(sorted_values)

    if num_observations % 2 != 0:
        median_index = num_observations // 2
        median = sorted_values[median_index]
    
    else:
        upper_median_index = num_observations // 2
        lower_median_index = upper_median_index - 1
        median = (sorted_values[lower_median_index] + sorted_values[upper_median_index]) / 2
    
    return median

median = calculate_median(dice_df)
print("Median:", median)
```

#### 4.- Define a function to calculate the four quartiles. You can use the function you defined above to compute the median but you are not allowed to use any methods or functions that directly calculate the quartiles. 

```python
def calculate_quartiles(df):
    sorted_values = sorted(df['Roll'])
    num_observations = len(sorted_values)

    median = calculate_median(df)
    
    if num_observations % 2 != 0:
        lower_half_end_index = num_observations // 2
        upper_half_start_index = lower_half_end_index + 1
    else:
        lower_half_end_index = num_observations // 2
        upper_half_start_index = lower_half_end_index

    lower_half = sorted_values[:lower_half_end_index]
    upper_half = sorted_values[upper_half_start_index:]

    q1 = calculate_median(pd.DataFrame({'Roll': lower_half}))
    q3 = calculate_median(pd.DataFrame({'Roll': upper_half}))

    return q1, median, q3

q1, median, q3 = calculate_quartiles(dice_df)
print("Q1:", q1)
print("Median (Q2):", median)
print("Q3:", q3)
```

#### 2.- Using the functions you defined in *challenge 2*, calculate the mean value of the hundred dice rolls.

```python
def calculate_mean(series):
    total_sum = 0
    num_observations = len(series)

    for value in series:
        total_sum += value

    mean = total_sum / num_observations
    return mean

mean_value = calculate_mean(roll_the_dice['value'])
print("Mean:", mean_value)
```

#### 4.- Plot the histogram. What do you see (shape, values...) ? How can you connect the mean value to the histogram?

```python
"""
the x-axis represents the dice values and the y-axis represents the frequency of each value.
the histogram might show a relatively uniform distribution if the dice rolls are fair, 
meaning each value occurs with approximately the same frequency. 
The mean value, calculated earlier, represents the average value of the dice rolls, 
which should fall somewhere in the center of the distribution if the rolls are fair.
"""
```
